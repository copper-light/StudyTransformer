```python
def check_script(char):
    code = ord(char)
    if 0xAC00 <= code <= 0xD7A3:
        return "Korean (Hangul)"
    elif 0x3040 <= code <= 0x309F:
        return "Japanese (Hiragana)"
    elif 0x4E00 <= code <= 0x9FFF:
        return "CJK (East Asia)" # 한/중/일 누가 썼는지 모름!
    elif 0x0041 <= code <= 0x005A:
        return "Latin (Unknown Nation)"
    else:
        return "Other"

print(check_script("가")) # Korean
print(check_script("A"))  # Latin (미국인지 프랑스인지 모름)
```
