```python
import csv

texts = []
with open('data/chatbot_kor/chatbot.csv', 'r', encoding='utf-8') as f:
    data = csv.reader(f)
    first_row = next(data)
    index = 0
    for row in data:

        with open(f'data/chatbot_kor/corpus/{index}.txt', 'w', encoding='utf-8') as wf:
            wf.write(row[0])
            index += 1

        with open(f'data/chatbot_kor/corpus/{index}.txt', 'w', encoding='utf-8') as wf:
            wf.write(row[1])
            index += 1
len(texts)
```

```python
import glob
```

```python
paths = [str(x) for x in glob.glob("./data/chatbot_kor/corpus/*.txt")]
```

```python
from tokenizers import SentencePieceBPETokenizer

tokenizer = SentencePieceBPETokenizer()
tokenizer.train(paths,  vocab_size=100000, min_frequency=5, show_progress=True, special_tokens=['<unk>', '<pad>', '<bos>', '<eos>', '<mask>', '<sep>'])

encoded = tokenizer.encode("이 영화는 정말 재미있습니다.")
print(encoded.tokens)
tokenizer.save("./voca.json",True)
```
